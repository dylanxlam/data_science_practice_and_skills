We will be considering the actual elapsed flight time "ActualElapsedTime" and the travel distance "Distance".

Create a simple linear regression model between these two values for predicting "ActualElapsedTime" using the "Distance" as the explanatory variable. 

Remove flights that do not have an "ActualElapsedTime" or "Distance" value, or their values are NaN or Infinity. 

Provide your Pandas code. 
What is the correlation coefficient between elapsed time and travel distance?
Provide your equation for this linear regression. 
Describe how well your model fits the data.
Provide the R-Squared of your model.

```python
from sklearn.metrics import r2_score

cleaned = flights.dropna(subset=['Distance', 'ActualElapsedTime'])

X = cleaned[['Distance']]
y = cleaned['ActualElapsedTime']

model = LinearRegression()
model.fit(X, y)

print('Correlation Coefficient:', cleaned['Distance'].corr(cleaned['ActualElapsedTime']))

print('Equation: Actual Elapsed Time =', model.coef_[0], '* (Distance) + ', model.intercept_)

y_pred = model.predict(X)


print('R-Squared Score:', r2_score(y, y_pred))
```

In summary, an R-squared score of 0.935 indicates a very strong fit of the regression model to the data, suggesting that the model explains a large proportion of the variability in the dependent variable and has strong predictive power.

```python
# Given data
x = [1, 2, 3, 4, 5, 6]
y = [101, 111, 122, 133, 144, 155]

# Initial parameters
intercept = 50
slope = .2

# Learning rate
learning_rate = 0.04

# Number of data points
n = len(x)

# Perform one iteration of gradient descent
for i in range(n):
    # Compute predicted value for the ith data point
    y_pred_i = intercept + slope * x[i]
    
    # Compute the gradients of the loss function with respect to intercept and slope
    grad_intercept = -2 * (y[i] - y_pred_i)  # Partial derivative of loss w.r.t. intercept
    grad_slope = -2 * x[i] * (y[i] - y_pred_i)  # Partial derivative of loss w.r.t. slope
    
    # Update parameters using the gradients and learning rate
    intercept -= learning_rate * grad_intercept
    slope -= learning_rate * grad_slope

# Output the updated parameters
print("Updated intercept:", intercept)
print("Updated slope:", slope)
```

```python
X = np.array([1,2,3,4,5,6])
y = np.array([101, 111, 122, 133, 144, 155])

learning_rate = 0.04
max_itr = 1
precision = 0.01

m = .2
b = 50

n = len(X)

cost_list = []
itr_list = []


for i in range(max_itr):
    prediction = b + m * X
    cost = (1/n) * sum(( prediction - y)**2)
    m_gradient = (1.0/n) * np.sum(X * (prediction - y)) 
    b_gradient = (1.0/n) * np.sum(prediction - y)
    m = m - learning_rate * m_gradient
    b = b - learning_rate * b_gradient

    cost_list.append(cost)
    itr_list.append(i+1)
    print('Iteration:', i+1, '; Cost:', cost)

    if len(cost_list) > 1 and abs(cost_list[-1] - cost_list[-2]) <= precision:
        print("Stoped at iteration", i+1)
        break

    if len(itr_list) == max_itr:
        print("Stoped at iteration", i+1)

print('Coefficients:')
print('Slope:', m)
print('Intercept:', b)
```
